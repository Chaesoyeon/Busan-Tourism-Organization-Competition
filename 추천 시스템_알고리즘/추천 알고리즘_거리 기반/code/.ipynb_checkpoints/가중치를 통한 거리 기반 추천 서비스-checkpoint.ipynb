```python
### import package

import pandas as pd
```

```python
df = pd.read_csv("../data/추천알고리즘.csv", encoding = "cp949")
```

```python
df.columns
```

```python
target = df[["시장명", "가게명", "가게 종류", "지역구", "주자창 여부", "지금 몇 명 가는지", "인스타 게시물 수", 'dist_현재위치', 'dist_추천식당', 'dist_추천장소', "별점"]]
```

```python
list_주차장 = []
for i in range(0, len(target),1): 
    if target["주자창 여부"][i] == "유":
        list_주차장.append(1)
    else:
        list_주차장.append(0)
target["주차장 존재"] = list_주차장
```

```python
target.columns
```

```python
## 스케일링을 먼저 진행하자

from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler_ = MinMaxScaler()
scaler_.fit(target[['지금 몇 명 가는지', '인스타 게시물 수',
       'dist_현재위치', 'dist_추천식당', 'dist_추천장소', '별점', '주차장 존재', 'non_car_score',
       'use_car_score']])
result = scaler_.transform(target[['지금 몇 명 가는지', '인스타 게시물 수',
       'dist_현재위치', 'dist_추천식당', 'dist_추천장소', '별점', '주차장 존재', 'non_car_score',
       'use_car_score']])
result_df = pd.DataFrame(result, columns = ['지금 몇 명 가는지', '인스타 게시물 수',
       'dist_현재위치', 'dist_추천식당', 'dist_추천장소', '별점', '주차장 존재', 'non_car_score',
       'use_car_score'])


### 데이터 프레임 합치기

target = pd.concat([target[['시장명', '가게명', '가게 종류', '지역구', '주자창 여부']], result_df], axis = 1)
```

```python
target
```

```python
### 대중교통 이용
# 가중치를 매겨보자!!
"""
주차장 여부 : 0
인스타 게시물 수 : 0.4
dist 현재위치: 0.6
dist 추천식당: 0.4
dist_추천장소: 0.4
지금 몇명 가는지: 0.6
별점: 0.5
"""

def non_car(data, 게시물, 현재위치, 추천식당, 추천장소, 지금몇명, 별점):
    list_score = []
    for i in range (0, len(data), 1):
        score = 게시물*data["인스타 게시물 수"][i] + 현재위치*data["dist_현재위치"][i] + 추천식당*data["dist_추천식당"][i] + 추천장소*data["dist_추천장소"][i]+ 지금몇명*data["지금 몇 명 가는지"][i] + 별점*data["별점"][i]
        list_score.append(score)
        
    data["non_car_score"] = list_score



### 자차 이용
def use_car(data, 주차장여부, 게시물, 현재위치, 추천식당, 추천장소, 지금몇명, 별점):
    list_score = []
    for i in range (0, len(data), 1):
        score = 주차장여부*data["주차장 존재"][i] + 게시물*data["인스타 게시물 수"][i] + 현재위치*data["dist_현재위치"][i] + 추천식당*data["dist_추천식당"][i] + 추천장소*data["dist_추천장소"][i] 
        + 지금몇명*data["지금 몇 명 가는지"][i] + 별점*data["별점"][i]
        list_score.append(score)
        
    data["use_car_score"] = list_score
```

```python
non_car(target, 0.4, 0.6, 0.4, 0.4, 0.6, 0.5)
```

```python
target.nlargest(5, "non_car_score", keep = "all")
```

```python
use_car(target, 1, 0.4, 0.6, 0.4, 0.4, 0.6, 0.5)
```

```python
target.nlargest(5, "use_car_score", keep = "all")
```
